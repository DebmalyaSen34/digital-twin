```python
def split_locus_name(locus_name):
    if 'and' in locus_name:
        genes_array = locus_name.split('and')
        genes_array = [gene.strip() for gene in genes_array]
        return genes_array
    else:
        return [locus_name.strip()]
```

## Visualization

```python
data_expanded = pd.read_csv("../data/pathway_gene_mapping.csv")
```

```python
import networkx as nx
from pyvis.network import Network
```

```python
G = nx.Graph()
```

```python
for index, row in data_expanded.iterrows():
    gene = row['Gene'].strip()
    pathway = row['Pathway']

    if pd.isna(gene) or pd.isna(pathway):
        continue

    gene = str(gene).strip()
    pathway = str(pathway).strip()

    G.add_node(gene, group='Gene', title=gene)
    G.add_node(pathway, group='Pathway', title=pathway)
    G.add_edge(gene, pathway)
```

```python
net = Network(notebook=True, height='750px', width='100%', bgcolor='#222222', font_color='white')

net.from_nx(G)
```

```python
for node in net.nodes:
    if node['group'] == 'Gene':
        node['color'] = '#00ff41' # Matrix Green for genes
        node['size'] = 10
    else:
        node['color'] = '#ff0044' # Red for pathways
        node['size'] = 25 # Make pathways larger hubs
```

```python
net.show("iPS189_Network.html")
```

```python
import re
```

```python
def extract_metabolites(equation_str):
    if pd.isna(equation_str):
        return []
    
    clean_eq = re.sub(r'\[.*?\]', '', equation_str)

    clean_eq = re.sub(r'\(\d+\)', '', clean_eq)

    clean_eq = clean_eq.replace('<==>', '+').replace('-->', '+').replace(':', '')

    compounds = clean_eq.split('+')

    clean_compounds = [comp.strip() for comp in compounds if comp.strip()]

    return list(set(clean_compounds))
```

```python
test_metabolites = extract_metabolites("acgam[e] + pep[c] --> acgam6p[c] + pyr[c]")
```

```python
for metabolite in test_metabolites:
    if metabolite not in metabolites_data['abbreviation'].values:
        print(f"Metabolite '{metabolite}' not found in metabolites dataset.")
```

```python
data['Metabolites'] = data['Equation'].apply(extract_metabolites)
```

## Simulation

```python
import cobra
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# --- 1. SETUP & LOADING ---
print("Loading model iPS189...")
try:
    model = cobra.io.read_sbml_model("iPS189.xml")
except OSError:
    print("Error: iPS189.xml not found. Please ensure the file is in the directory.")
    exit()

# Set the objective to Biomass (Growth)
# Note: We use the ID 'R_Biomass' found in your XML file
model.objective = "Biomass"

# --- 2. BASELINE SIMULATION (Wild Type) ---
print("\nRunning Wild-Type Simulation...")
solution_wt = model.optimize()

if solution_wt.status != 'optimal':
    print("Optimization failed! Check model constraints.")
else:
    print(f"Wild-Type Growth Rate: {solution_wt.objective_value:.4f} 1/h")
    print(f"Predicted Doubling Time: {(np.log(2) / solution_wt.objective_value) * 60:.2f} min")

# --- 3. SIMULATE DRUG RESPONSE (Dose-Response Curve) ---
# We will simulate a drug that inhibits gene 'MG041' (Glucose Transport)
# This gene is essential for carbon uptake in this organism.
drug_target_gene = "MG041"
print(f"\nSimulating Drug Response targeting {drug_target_gene} (Glucose Transport)...")

dose_levels = np.linspace(0, 1.0, 20)  # 0% to 100% Inhibition
growth_rates = []
flux_data = []

# Get the reaction associated with the target gene
# We find reactions that require this gene to function
target_reactions = [rxn for rxn in model.reactions if drug_target_gene in rxn.gene_reaction_rule]

# Store original bounds to reset later
original_bounds = {rxn.id: rxn.bounds for rxn in target_reactions}

for inhibition in dose_levels:
    # Apply inhibition: Reduce the flux bound of the target reaction(s)
    # 0.0 = No Drug (100% activity), 1.0 = Full Drug (0% activity)
    remaining_activity = 1.0 - inhibition
    
    for rxn in target_reactions:
        # We assume the drug affects both directions of a reversible reaction equally
        lower, upper = original_bounds[rxn.id]
        rxn.bounds = (lower * remaining_activity, upper * remaining_activity)
    
    # Run Simulation
    sol = model.optimize()
    
    # Record Data
    growth = sol.objective_value if sol.status == 'optimal' else 0.0
    growth_rates.append(growth)
    
    # Capture fluxes of key pathways for visualization
    # We'll grab Glycolysis (PGI), Biomass, and Transport
    flux_snapshot = {
        'Growth': growth,
        'Glucose_Transport': model.reactions.get_by_id('GLCpts').flux if 'GLCpts' in model.reactions else 0,
        'Glycolysis_PGI': model.reactions.get_by_id('PGI').flux if 'PGI' in model.reactions else 0,
        'ATP_Synthase': model.reactions.get_by_id('ATPS4r').flux if 'ATPS4r' in model.reactions else 0
    }
    flux_data.append(flux_snapshot)

# Reset bounds for safety
for rxn in target_reactions:
    rxn.bounds = original_bounds[rxn.id]

# --- 4. VISUALIZE RESULTS ---
# Plot 1: Drug Dose-Response Curve
plt.figure(figsize=(10, 6))
plt.plot(dose_levels * 100, growth_rates, 'r-o', linewidth=2, label=f'Target: {drug_target_gene}')
plt.axhline(y=solution_wt.objective_value, color='gray', linestyle='--', label='Wild Type')
plt.title(f'Drug Dose-Response: Inhibition of {drug_target_gene}', fontsize=14)
plt.xlabel('Drug Concentration (% Inhibition)', fontsize=12)
plt.ylabel('Growth Rate (1/h)', fontsize=12)
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.savefig('drug_response_curve.png')
print("Saved plot: drug_response_curve.png")

# Plot 2: Metabolic Flux Response (Heatmap)
# Convert flux data to DataFrame
flux_df = pd.DataFrame(flux_data)
flux_df.index = [f"{int(d*100)}%" for d in dose_levels] # Use dose as index

# Normalize data for better heatmap visualization (0 to 1 scale per reaction)
flux_normalized = flux_df / flux_df.max()

plt.figure(figsize=(8, 8))
sns.heatmap(flux_normalized, cmap="viridis", annot=False)
plt.title('Metabolic Pathway Response to Drug', fontsize=14)
plt.ylabel('Drug Dosage (% Inhibition)')
plt.xlabel('Metabolic Fluxes')
plt.tight_layout()
plt.savefig('gene_response_heatmap.png')
print("Saved plot: gene_response_heatmap.png")

plt.show()
```
